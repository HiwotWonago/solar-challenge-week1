```python
# Average rainfall per region per year
annual_rainfall = df.groupby(['Country', 'Region', 'Year'])['Rainfall_mm'].mean().reset_index()

# Pivot for plotting
pivot_df = annual_rainfall.pivot_table(index='Year', columns='Region', values='Rainfall_mm')
print(pivot_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_df)
plt.title('Average Annual Rainfall (mm) per Region: 2018‚Äì2022')
plt.ylabel('Rainfall (mm)')
plt.xlabel('Year')
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
